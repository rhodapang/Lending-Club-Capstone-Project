```python
# Import two datasets
colnames = ["LOAN_ID", "ORIG_CHN", "Seller.Name", "ORIG_RT", "ORIG_AMT", "ORIG_TRM", "ORIG_DTE","FRST_DTE", "OLTV", 
            "OCLTV", "NUM_BO", "DTI", "CSCORE_B", "FTHB_FLG", "PURPOSE", "PROP_TYP","NUM_UNIT", "OCC_STAT", 
            "STATE", "ZIP_3", "MI_PCT", "Product.Type", "CSCORE_C", "MI_TYPE", "RELOCATION_FLG"]
Data_A = pd.read_table('D:/1-Intern/Data/2016Q3/Acquisition_2016Q3.txt', 
                       sep = '|', header = None, names = colnames)
colnames_P = ["LOAN_ID", "Monthly.Rpt.Prd", "Servicer.Name", "LAST_RT", "LAST_UPB", "Loan.Age", "Months.To.Legal.Mat", 
              "Adj.Month.To.Mat", "Maturity.Date", "MSA", "Delq.Status", "MOD_FLAG", "Zero.Bal.Code", 
              "ZB_DTE", "LPI_DTE", "FCC_DTE","DISP_DT", "FCC_COST", "PP_COST", "AR_COST", "IE_COST", "TAX_COST", "NS_PROCS",
              "CE_PROCS", "RMW_PROCS", "O_PROCS", "NON_INT_UPB", "PRIN_FORG_UPB_FHFA", "REPCH_FLAG", "PRIN_FORG_UPB_OTH", "TRANSFER_FLG"]
Data_P = pd.read_table('D:/1-Intern/Data/2016Q3/Performance_2016Q3.txt', 
                       sep = '|', header = None, names = colnames_P,low_memory=False)
```

```python
# Convert string into datetime 
from datetime import datetime
Data_P['Monthly.Rpt.Prd']= [datetime.strptime(x, '%m/%d/%Y') for x in Data_P.loc[:,'Monthly.Rpt.Prd']]
```

```python
df['UPB_From_Last_Month']=df['LAST_UPB'].groupby(df['LOAN_ID']).shift(1)
```

```python

df['Monthly.Rpt.Prd']= [datetime.strptime(x, '%m/%d/%Y') for x in df.loc[:,'Monthly.Rpt.Prd']]
```

```python
df['UPB_From_Last_Month']
```

```python
# Define a function to calculate origination monthly payment  
def Monthly_Payment_Calculation(x, y,z):
    return (x * (y / 1200)* ((1 + (y / 1200))**z)) / (((1+( y/1200))**z) -1)

df['Montly_Payment'] = round(Monthly_Payment_Calculation(df.loc[:, 'ORIG_AMT'], df.loc[:, 'ORIG_RT'], df.loc[:, 'ORIG_TRM']),2)
df['Montly_Payment']
```

```python
#Calculate monthly interest should be paid
df['Monthly_Interest_Payment']= df['UPB_From_Last_Month'] * (df['LAST_RT']/1200)
df['Monthly_Interest_Payment']
```

```python
#Calculate monthly principal should be paid
df['Scheduled_Monthly_Principal_Payment']= df['Montly_Payment']- df['Monthly_Interest_Payment']
df['Scheduled_Monthly_Principal_Payment']
```

```python
# How many principal of loan  has been deducted. 
#'UPB_From_Last_Month' is the UPB in the beginning of the month and 'LAST_UPB' is the UPB at the end of the month
df['Actual_Loan_Amount_Deducted'] = df['UPB_From_Last_Month']- df['LAST_UPB']
df['Actual_Loan_Amount_Deducted'] 
```

```python
# Calculate the total loan payment made 
df['Actual_Monthly_Payment'] = df['Actual_Loan_Amount_Deducted']+ df['Monthly_Interest_Payment']
df['Actual_Monthly_Payment']
```

```python
# Before calculating montly SMM and CPR, loans should be grouped by monthly. 
Monthly_Total_Payment = df[['Actual_Monthly_Payment','Montly_Payment','UPB_From_Last_Month', 'LAST_UPB', 'Scheduled_Monthly_Principal_Payment']].groupby(df['Monthly.Rpt.Prd']).sum()
   
```

```python
# For the loan is prepaid ('Zero.Bal.Code'] == 1.0), it shows that 'Actual_Loan_Amount_Deducted' = 0, 
# so 'Actual_Monthly_Payment' of those loan should be calculate specially. 
Loan_Prepay = df[df['Zero.Bal.Code']==1.0]
```

```python
Loan_Prepay.shape
```

```python
df_1 = Loan_Prepay['UPB_From_Last_Month'] .groupby(df['Monthly.Rpt.Prd']).sum()
```

```python
df_1 = pd.DataFrame(df_1)
df_1
```

```python
df_1.columns
```

```python
Monthly_Total_Payment['Actual_Monthly_Payment_1']= Monthly_Total_Payment['Actual_Monthly_Payment'] + df_1['UPB_From_Last_Month']
```

```python
def SMM(u,v,x,y):
    return(u - v)/(x - y)

Monthly_Total_Payment['SMM'] = round( SMM(Monthly_Total_Payment.loc[:, 'Actual_Monthly_Payment_1'], Monthly_Total_Payment.loc[:, 'Montly_Payment'],
                                Monthly_Total_Payment.loc[:, 'UPB_From_Last_Month'], Monthly_Total_Payment.loc[:, 'Scheduled_Monthly_Principal_Payment']),8)
```

```python
#calculate CPR 
def CPR(x):
    return(1-(1-x)**12)

Monthly_Total_Payment['CPR']= round(CPR(Monthly_Total_Payment.loc[:, 'SMM']), 6) 
```
